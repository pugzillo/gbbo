```python
from bs4 import BeautifulSoup
import requests
import numpy as np
import os
import pandas as pd
```

```python
url = os.environ["GBBO_RECIPE_URL"]
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
# Get recipe name
soup.find(class_= 'page-banner__title').h1.contents
```

```python
# Get baker's name
soup.find(class_='recipe-introduction__baker__name').a.contents
```

```python
# Get recipe details
for string in soup.find_all(class_='recipe-details__item'):
    print(string.find(class_= 'label').contents)
    print(string.find(class_='value').contents)
```

```python
# Get recipe ingredients
for string in soup.find_all(class_ = 'recipe-sidebar__section recipe-sidebar__section--ingredients'):
    for item in string.find_all('p'):
        print(item.contents)
```

```python
# get the recipe steps
for string in soup.find_all(class_='recipe-instructions'):
    print(len(string.find_all('p'))) # number of steps
    for step in string.find_all('p'):
        print(step.contents[-1])
```
